```python
import pandas as pd
dados = pd.read_csv("dados/aluguel.csv", sep =";")
#Para criar uma nova coluna(variável), vai ser criado a variável "Valor Bruto"
dados["Valor Bruto"] = dados["Valor"] + dados["Condominio"] + dados["IPTU"]
dados
```

```python
#Criando a variável "Valor m2". O "round" arendonda os valores
dados["Valor m2"] = (dados["Valor Bruto"] / dados["Area"]).round(2)
dados
```

```python
#Criando  variável com a função lambda. O "apply" aplica uma função a cada linha. Ele vai analisar sua condição e aplicá-la
casa = ["Casa", "Casa de condomínio", "Casa de vila"]
dados["Tipo agregado"] = dados["Tipo"].apply(lambda x: "Casa" if x in casa else "Apartamento")
dados
```

## Excluindo as variáveis(colunas)

```python
dados_aux = pd.DataFrame(dados[["Tipo agregado", "Valor m2", "Valor Bruto"]])
dados_aux
```

```python
# uma das possibilidades é usar o "del"
del dados_aux["Valor m2"]
dados_aux.head(10)
```

```python
# Outra possibilidade é usar o "pop", ele mostra a coluna que foi excluída
dados_aux.pop("Valor Bruto")
```

```python
dados_aux.head(10)
```

```python
# Usando o "drop" você poderá excluir a coluna tudo de uma vez só. Para isso eu preciso informar o eixo "axis" e o inplace.
# axis = 1 == Coluna    axis = 0 == Linha
dados.drop(["Valor Bruto"], axis = 1, inplace = True)
dados.head(10)
```

```python
dados.to_csv("dados/aluguel_residencial.csv", sep = ";", index = False)
```
