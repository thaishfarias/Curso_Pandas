```python
import pandas as pd
dados = pd.read_csv("dados/aluguel.csv", sep = ";")
residencial = ['Quitinete',
 'Casa',
 'Conjunto Comercial/Sala',
 'Apartamento',
 'Casa de Condomínio']
```

```python
selecao =  dados["Tipo"].isin(residencial)
```

```python
contagem_de_quartos = dados["Tipo"].value_counts().to_frame().reset_index()
contagem_de_quartos.columns = ["Tipo", "total"]
contagem_de_quartos.head()

import seaborn as sns

sns.catplot(y = "Tipo", kind = "count", data = contagem_de_quartos, aspect = 4)
```
